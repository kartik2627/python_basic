```python
####matrix plot
### heat map and cluster map
gap = px.data.gapminder()
gap
```

```python
temp_df = gap.pivot(index='country',columns = 'year',values='lifeExp')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=temp_df)
```

```python
temp_gap=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(temp_gap,annot=True)
```
