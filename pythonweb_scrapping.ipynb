```python
import requests
wiki_link ="https://en.wikipedia.org/wiki/Nymphaeaceae"
link = requests.get(wiki_link).text
```

```python
print(link)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(link, 'lxml')
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.string
```

```python
soup.a
```

```python
soup.find_all("a")
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
all_tables = soup.find_all('table')
print(all_tables)
```
